```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
import warnings

warnings.filterwarnings("ignore")

def train_statsmodels_arima_on_all_columns(csv_path, ticker, steps=30, order=(6, 1, 3)):
    print(f"\n=== Training ARIMA models for {ticker} ===")

    df = pd.read_csv(csv_path, parse_dates=["Date"])
    df.set_index("Date", inplace=True)

    for column in df.columns:
        print(f"\n--- Column: {column} ---")
        series = df[column].dropna()

        # Train-test split
        n = int(len(series) * 0.8)
        train = series[:n]
        test = series[n:]

        # Fit ARIMA model
        model = ARIMA(train, order=order)
        result = model.fit()

        # Forecast
        forecast = result.forecast(steps=steps)
        forecast_index = test[:steps].index

        # Calculate RMSE
        rmse = mean_squared_error(test[:steps], forecast, squared=False)
        print(f"RMSE: {rmse:.4f}")

        # Plot
        plt.figure(figsize=(10, 4))
        plt.plot(test[:steps], label="Actual")
        plt.plot(forecast_index, forecast, label="Forecast")
        plt.title(f"{ticker} - {column}: Forecast vs Actual")
        plt.legend()
        plt.tight_layout()
        plt.show()

# Example usage
tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS', 'TSLA']
for ticker in tickers:
    train_statsmodels_arima_on_all_columns(f"{ticker}_data.csv", ticker)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
import warnings
import os

warnings.filterwarnings("ignore")

def train_statsmodels_arima_on_all_columns(csv_path, ticker, steps=30, order=(6, 1, 3), results=[]):
    print(f"\n=== Training ARIMA models for {ticker} ===")

    df = pd.read_csv(csv_path, parse_dates=["Date"])
    df.set_index("Date", inplace=True)

    output_dir = "forecast_plots"
    os.makedirs(output_dir, exist_ok=True)

    for column in df.columns:
        print(f"\n--- Column: {column} ---")
        series = df[column].dropna()

        # Train-test split
        n = int(len(series) * 0.8)
        train = series[:n]
        test = series[n:]

        if len(test) < steps:
            print("❌ Not enough test data for forecasting.")
            continue

        # Fit ARIMA model
        try:
            model = ARIMA(train, order=order)
            result = model.fit()

            # Forecast
            forecast = result.forecast(steps=steps)
            forecast_index = test[:steps].index

            # Calculate RMSE
            rmse = mean_squared_error(test[:steps], forecast, squared=False)
            print(f"RMSE: {rmse:.4f}")

            # Save results
            results.append({
                "Ticker": ticker,
                "Column": column,
                "RMSE": round(rmse, 4)
            })

            # Plot and save
            plt.figure(figsize=(10, 4))
            plt.plot(test[:steps], label="Actual")
            plt.plot(forecast_index, forecast, label="Forecast")
            plt.title(f"{ticker} - {column}: Forecast vs Actual")
            plt.legend()
            plt.tight_layout()

            plot_filename = f"{output_dir}/{ticker}_{column}_forecast.png"
            plt.savefig(plot_filename)
            plt.close()
        except Exception as e:
            print(f"⚠️ Error training {ticker} - {column}: {e}")

    return results


# Run training for all tickers and collect results
tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS', 'TSLA']
all_results = []

for ticker in tickers:
    all_results = train_statsmodels_arima_on_all_columns(f"{ticker}_data.csv", ticker, results=all_results)

# Save RMSE results to CSV
rmse_df = pd.DataFrame(all_results)
rmse_df.to_csv("arima_rmse_results.csv", index=False)
print("\n✅ RMSE results saved to arima_rmse_results.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
from sklearn.metrics import mean_squared_error
import seaborn as sns
import numpy as np
import warnings
import os

warnings.filterwarnings("ignore")

def plot_residual_diagnostics(residuals, title_prefix, output_dir):
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    
    # Residual histogram
    sns.histplot(residuals, kde=True, ax=axes[0])
    axes[0].set_title(f"{title_prefix} - Residual Histogram")

    # ACF of residuals
    from statsmodels.graphics.tsaplots import plot_acf
    plot_acf(residuals, ax=axes[1], lags=30)
    axes[1].set_title(f"{title_prefix} - Residual ACF")

    plt.tight_layout()
    file_path = os.path.join(output_dir, f"{title_prefix}_residual_diagnostics.png")
    plt.savefig(file_path)
    plt.close()


def train_arima_with_validation(csv_path, ticker, steps=30, order=(6, 1, 3), results=[]):
    print(f"\n=== Validating ARIMA for {ticker} ===")
    df = pd.read_csv(csv_path, parse_dates=["Date"])
    df.set_index("Date", inplace=True)

    output_dir = "forecast_plots"
    os.makedirs(output_dir, exist_ok=True)

    for column in df.columns:
        print(f"\n--- Column: {column} ---")
        series = df[column].dropna()
        n = int(len(series) * 0.8)
        train = series[:n]
        test = series[n:]

        if len(test) < steps:
            print("❌ Not enough test data.")
            continue

        try:
            # Fit ARIMA
            model = ARIMA(train, order=order)
            result = model.fit()

            # Forecast
            forecast = result.forecast(steps=steps)
            forecast_index = test[:steps].index
            rmse = mean_squared_error(test[:steps], forecast, squared=False)

            # Residual Diagnostics
            residuals = result.resid
            plot_residual_diagnostics(residuals, f"{ticker}_{column}", output_dir)

            # Ljung-Box test
            lb_test = acorr_ljungbox(residuals, lags=[10], return_df=True)
            lb_p = lb_test['lb_pvalue'].values[0]
            lb_result = "OK" if lb_p > 0.05 else "Autocorrelated"

            # AIC/BIC
            aic = result.aic
            bic = result.bic

            print(f"RMSE: {rmse:.4f} | Ljung-Box P-Value: {lb_p:.4f} ({lb_result}) | AIC: {aic:.2f} | BIC: {bic:.2f}")

            results.append({
                "Ticker": ticker,
                "Column": column,
                "RMSE": round(rmse, 4),
                "LjungBox_p": round(lb_p, 4),
                "AIC": round(aic, 2),
                "BIC": round(bic, 2),
                "ResidualStatus": lb_result
            })

            # Plot forecast
            plt.figure(figsize=(10, 4))
            plt.plot(test[:steps], label="Actual")
            plt.plot(forecast_index, forecast, label="Forecast")
            plt.title(f"{ticker} - {column}: Forecast vs Actual")
            plt.legend()
            plt.tight_layout()
            plt.savefig(os.path.join(output_dir, f"{ticker}_{column}_forecast.png"))
            plt.close()

        except Exception as e:
            print(f"⚠️ Error for {ticker} - {column}: {e}")

    return results


# Run across all tickers
tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS', 'TSLA']
all_results = []

for ticker in tickers:
    all_results = train_arima_with_validation(f"{ticker}_data.csv", ticker, results=all_results)

# Save summary
results_df = pd.DataFrame(all_results)
results_df.to_csv("arima_validation_results.csv", index=False)
print("\n✅ All validation results saved to arima_validation_results.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Set random seed for reproducibility
np.random.seed(42)

# Create sequences from time series
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

# Train LSTM for a single ticker
def train_lstm_for_ticker(filename, ticker, results, seq_length=30):
    print(f"\nProcessing: {ticker}")
    
    if not os.path.exists(filename):
        print(f"File {filename} not found.")
        return results

    # Load data
    df = pd.read_csv(filename, parse_dates=['Date'], index_col='Date')
    
    if 'Close' not in df.columns:
        print(f"'Close' column not found in {filename}. Skipping.")
        return results

    # Sort just in case
    df = df.sort_index()
    data = df['Close'].values.reshape(-1, 1)

    # Scale data
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(data)

    # Create sequences
    X, y = create_sequences(scaled, seq_length)
    if len(X) == 0:
        print(f"Not enough data to form sequences for {ticker}")
        return results

    # Train-test split (80-20)
    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Reshape for LSTM
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

    # Build model
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(seq_length, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')

    # Train
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)

    # Predict
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)
    y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Evaluate
    rmse = mean_squared_error(y_test_rescaled, predictions, squared=False)
    results.append({'Ticker': ticker, 'RMSE': rmse})

    # Optional Plot
    plt.figure(figsize=(10, 4))
    plt.plot(y_test_rescaled, label='Actual')
    plt.plot(predictions, label='Predicted')
    plt.title(f"LSTM Forecast: {ticker}")
    plt.legend()
    plt.show()

    return results

# Run across all tickers
tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS', 'TSLA']
all_results = []

for ticker in tickers:
    all_results = train_lstm_for_ticker(f"{ticker}_data.csv", ticker, results=all_results)

# Final results table
print("\nPerformance Summary:")
results_df = pd.DataFrame(all_results)
print(results_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import os

# Create lag features for time series
def create_lag_features(data, n_lags=5):
    df = pd.DataFrame(data, columns=['Close'])
    for i in range(1, n_lags + 1):
        df[f'lag_{i}'] = df['Close'].shift(i)
    df.dropna(inplace=True)
    return df

# Train Random Forest for a single ticker
def train_rf_for_ticker(filename, ticker, results, n_lags=5):
    print(f"\nProcessing: {ticker}")
    
    if not os.path.exists(filename):
        print(f"File {filename} not found.")
        return results

    # Load data
    df = pd.read_csv(filename, parse_dates=['Date'], index_col='Date')
    
    if 'Close' not in df.columns:
        print(f"'Close' column not found in {filename}. Skipping.")
        return results

    # Sort
    df = df.sort_index()
    close_data = df['Close'].values

    # Create features
    df_feat = create_lag_features(close_data, n_lags=n_lags)
    X = df_feat.drop('Close', axis=1).values
    y = df_feat['Close'].values

    # Train-test split
    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    if len(X_train) == 0 or len(X_test) == 0:
        print(f"Not enough data to train/test for {ticker}.")
        return results

    # Train model
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Predict and evaluate
    predictions = model.predict(X_test)
    rmse = mean_squared_error(y_test, predictions, squared=False)
    results.append({'Ticker': ticker, 'RMSE': rmse})

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(y_test, label='Actual')
    plt.plot(predictions, label='Predicted')
    plt.title(f"Random Forest Forecast: {ticker}")
    plt.legend()
    plt.show()

    return results

# Run for all tickers
tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS', 'TSLA']
all_results_rf = []

for ticker in tickers:
    all_results_rf = train_rf_for_ticker(f"{ticker}_data.csv", ticker, results=all_results_rf)

# Final results
print("\nRandom Forest RMSE Summary:")
results_df_rf = pd.DataFrame(all_results_rf)
print(results_df_rf)
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import os

def create_features(df, lags=[1, 2, 3], rolling_windows=[3, 5]):
    df = df.copy()
    for lag in lags:
        df[f'lag_{lag}'] = df['Close'].shift(lag)

    for window in rolling_windows:
        df[f'roll_mean_{window}'] = df['Close'].shift(1).rolling(window=window).mean()
        df[f'roll_std_{window}'] = df['Close'].shift(1).rolling(window=window).std()

    df['returns'] = df['Close'].pct_change()
    df = df.dropna()
    return df

def train_xgboost_model(csv_file, ticker):
    df = pd.read_csv(csv_file, parse_dates=['Date'])
    df = df.sort_values('Date')
    df = create_features(df)

    feature_cols = [col for col in df.columns if col not in ['Date', 'Close']]
    X = df[feature_cols]
    y = df['Close']

    # 80% train, 20% test
    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

    model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=4, random_state=42)
    model.fit(X_train, y_train)

    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, preds))

    return {'Ticker': ticker, 'RMSE': rmse}

# Run across all tickers
tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS', 'TSLA']
xgb_results = []

for ticker in tickers:
    filename = f"{ticker}_data.csv"
    if os.path.exists(filename):
        result = train_xgboost_model(filename, ticker)
        xgb_results.append(result)

# Show summary
results_df = pd.DataFrame(xgb_results)
print("\nXGBoost RMSE Summary:")
print(results_df.sort_values('RMSE'))
```

```python
import pandas as pd
import xgboost as xgb
import numpy as np
import pickle
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS', 'TSLA']
rmse_results = []

for ticker in tickers:
    if ticker == 'TSLA':
        continue  # Skip TSLA
    
    # Load dataset
    df = pd.read_csv(f'{ticker}_data.csv')
    
    # Convert the 'Date' column to datetime if it's not already
    if df['Date'].dtype == 'object':
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    
    # Drop the 'Date' column (if you don't need it as a feature)
    df = df.drop(columns=['Date'])

    df = df.dropna()  # Drop rows with missing values
    
    # Features and target
    X = df.drop(columns=['Close'])  # Drop the 'Close' column as target
    y = df['Close']
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
    
    # XGBoost model
    model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)
    model.fit(X_train, y_train)
    
    # Predict and evaluate
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    rmse_results.append({'Ticker': ticker, 'RMSE': rmse})
    
    # Export model (skip TSLA as mentioned)
    with open(f'{ticker}_xgb_model.pkl', 'wb') as f:
        pickle.dump(model, f)

# Print RMSE Summary
print("\nXGBoost RMSE Summary:")
print(pd.DataFrame(rmse_results))
```

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import xgboost as xgb
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping

# Configuration
sns.set(style='whitegrid')
tickers = ['AAPL', 'MSFT', 'GOOG', 'TCS.NS', 'TATAMOTORS.NS', 'NBCC.NS']
data_dir = 'D:/ML Folders/ml_env/GitHub/dava_project/Stock-Prediction-using-ARIMA/data'
os.makedirs('forecast_plots', exist_ok=True)

def prepare_data(df, window=10):
    """Preprocess data for time series forecasting"""
    df = df[['Close']].dropna().copy()
    scaler = MinMaxScaler()
    df['scaled'] = scaler.fit_transform(df[['Close']])
    X, y = [], []
    for i in range(window, len(df)):
        X.append(df['scaled'].values[i-window:i])
        y.append(df['scaled'].values[i])
    return np.array(X), np.array(y), scaler

def train_lstm(X_train, y_train, X_test, y_test, scaler):
    """Train LSTM model and return predictions"""
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(X_train.shape[1], 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=50, verbose=0, 
              callbacks=[EarlyStopping(monitor='loss', patience=5)])
    
    # Generate and inverse transform predictions
    test_pred = model.predict(X_test)
    train_pred = model.predict(X_train)
    predictions = np.concatenate([train_pred, test_pred])
    return model, scaler.inverse_transform(predictions)

def train_xgb(X, y, scaler):
    """Train XGBoost model and return predictions"""
    X_flat = X.reshape(X.shape[0], X.shape[1])
    model = xgb.XGBRegressor(objective='reg:squarederror')
    model.fit(X_flat, y)
    predictions = model.predict(X_flat)
    return model, scaler.inverse_transform(predictions.reshape(-1, 1))

def train_arima(df):
    """Train ARIMA model with stationarity check"""
    series = df['Close'].dropna()
    
    # Check stationarity
    result = adfuller(series)
    if result[1] > 0.05:
        d = 1
        print(f"Non-stationary series, differencing (d={d})")
    else:
        d = 0
        
    try:
        model = ARIMA(series, order=(5, d, 2))
        fitted = model.fit()
        return fitted, fitted.fittedvalues
    except Exception as e:
        print(f"ARIMA failed: {str(e)}")
        return None, None

# Visualization functions
def plot_predictions(ticker, actual, predictions, model_name):
    plt.figure(figsize=(12, 6))
    plt.plot(actual, label='Actual')
    plt.plot(predictions, alpha=0.7, label=f'{model_name} Predictions')
    plt.title(f'{ticker} Price Predictions')
    plt.legend()
    plt.savefig(f'forecast_plots/{ticker}_{model_name}_comparison.png')
    plt.close()

def plot_feature_importance(ticker, model, model_type):
    plt.figure(figsize=(10, 6))
    if model_type == 'xgb':
        xgb.plot_importance(model)
    plt.title(f'{ticker} {model_type} Feature Importance')
    plt.savefig(f'forecast_plots/{ticker}_{model_type}_importance.png')
    plt.close()

def plot_arima_diagnostics(ticker, results):
    if results is None: return
    
    plt.figure(figsize=(12, 8))
    # Residual plots
    plt.subplot(2,2,1)
    plt.plot(results.resid)
    plt.title('Residuals')
    
    plt.subplot(2,2,2)
    sns.histplot(results.resid, kde=True)
    
    plt.subplot(2,2,3)
    stats.probplot(results.resid, plot=plt)
    
    plt.subplot(2,2,4)
    plot_acf(results.resid, lags=20)
    
    plt.suptitle(f'{ticker} ARIMA Diagnostics')
    plt.savefig(f'forecast_plots/{ticker}_arima_diagnostics.png')
    plt.close()

# Main training loop
results = []
for ticker in tickers:
    file_path = os.path.join(data_dir, f'{ticker}_data.csv')
    if not os.path.exists(file_path):
        print(f"Missing {file_path}")
        continue

    try:
        df = pd.read_csv(file_path, parse_dates=['Date'], index_col='Date').sort_index()
        X, y, scaler = prepare_data(df)
        actual_prices = scaler.inverse_transform(y.reshape(-1, 1))
        
        # Split data
        split = int(0.8 * len(X))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]
        
        # LSTM Training
        X_train_lstm = X_train.reshape(*X_train.shape, 1)
        X_test_lstm = X_test.reshape(*X_test.shape, 1)
        lstm_model, lstm_pred = train_lstm(X_train_lstm, y_train, X_test_lstm, y_test, scaler)
        lstm_rmse = np.sqrt(mean_squared_error(actual_prices, lstm_pred))
        
        # XGBoost Training
        xgb_model, xgb_pred = train_xgb(X, y, scaler)
        xgb_rmse = np.sqrt(mean_squared_error(actual_prices, xgb_pred))
        
        # ARIMA Training
        arima_model, arima_pred = train_arima(df)
        arima_rmse = np.sqrt(mean_squared_error(df['Close'][1:], arima_pred[1:])) if arima_pred is not None else np.nan
        
        # Store results
        results.append({
            'Ticker': ticker,
            'LSTM_RMSE': lstm_rmse,
            'XGB_RMSE': xgb_rmse,
            'ARIMA_RMSE': arima_rmse
        })
        
        # Generate plots
        plot_predictions(ticker, df['Close'], lstm_pred, 'LSTM')
        plot_predictions(ticker, df['Close'], xgb_pred, 'XGBoost')
        plot_feature_importance(ticker, xgb_model, 'xgb')
        if arima_model:
            plot_arima_diagnostics(ticker, arima_model)
            plot_predictions(ticker, df['Close'], arima_pred, 'ARIMA')

    except Exception as e:
        print(f"Error processing {ticker}: {str(e)}")

# Save and display results
results_df = pd.DataFrame(results)
results_df.to_csv('model_performance.csv', index=False)
print("\nFinal Results:")
print(results_df)

# Comparison plot
plt.figure(figsize=(12, 6))
melted = results_df.melt(id_vars='Ticker', var_name='Model', value_name='RMSE')
sns.barplot(data=melted, x='Ticker', y='RMSE', hue='Model')
plt.title('Model Performance Comparison')
plt.savefig('forecast_plots/model_comparison.png')
plt.show()
```
